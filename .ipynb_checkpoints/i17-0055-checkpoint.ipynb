## Tokenization

### Text To Tokens

```python
def addStartEndSentenceMark(sentence):
    '''
    Sentence breaken into token as parameter
    '''
    if (type(sentence) == list):
        sentence.insert(0,'<s>')
        sentence.append('</s>')
    elif (type(sentence) == str):
        sentence="<s> "+sentence+" </s>"
    return sentence
```

```python
def applySentenceTokenization(sentence):
#     tokenize sentences into words
    tokens=nltk.tokenize.word_tokenize(sentence)
    return tokens
```

```python
def applyTextTokenization(text):

    #     tokenize text into sentences
    listSentences=nltk.tokenize.sent_tokenize(text)

    # tokenize into words
    listSentences=[applySentenceTokenization(sentence) for sentence in listSentences]    

    #     adding start & end sentence mark
    listSentences=[addStartEndSentenceMark(sentence) for sentence in listSentences]
    
#     listSentences=[TreebankWordDetokenizer().detokenize(sentence) for sentence in listSentences]
    return listSentences
```

### N-grams Tokenization

```python
# n-grams to count mapping

def ngramCalculater(data):
    trigrams_data = defaultdict(lambda :0)
    bigrams_data = defaultdict(lambda:0)
    unigrams_data = defaultdict(lambda:0)
    N=3

    for plot in data:

        for i in range(len(plot)):
    #         into plot[i] i.e sentences in plot
            sentence=plot[i]

    #         trigrams in a sentence i.e len(sent)-2
            trigrams= [sentence[j:j+N] for j in range(len(sentence)-N+1)]
    #         bigrams in a sentence i.e len(sent)-1
            bigrams= [sentence[j:j+N-1] for j in range(len(sentence)-(N-1)+1)]
    #         unigrams in a sentence 
            unigrams= [sentence[j] for j in range(len(sentence))]

            for gram in trigrams:
                gram=(tuple(gram))
                trigrams_data[gram]+=1
            for gram in bigrams:
                gram=(tuple(gram))
                bigrams_data[gram]+=1
            for gram in unigrams:
                unigrams_data[gram]+=1
    return trigrams_data,bigrams_data,unigrams_data
```

```python
books_trigrams,books_bigrams,books_unigrams=ngramCalculater(df_plotSummaryBooks['Plot'])
print("Books n-grams Details:\n - Trigrams:{}\n - Bigrams:{}\n - Unigrams{}".format(len(books_trigrams.keys()),len(books_bigrams.keys()),len(books_unigrams.keys())))
```

```python
movies_trigrams,movies_bigrams,movies_unigrams=ngramCalculater(df_plotSummaryMovies['Plot'])
print("Movies n-grams Details:\n - Trigrams:{}\n - Bigrams:{}\n - Unigrams{}".format(len(movies_trigrams.keys()),len(movies_bigrams.keys()),len(movies_unigrams.keys())))
```

### Text Generation

```python
def generateRandomSentenceBooks(text):
    maxTokens=100
    tokens=applySentenceTokenization(text)

#             end of sentence ; useful for recursion 
    if (len(tokens)!=0 and (tokens[len(tokens)-1]=='.' or len(tokens)>maxTokens)):
        text=TreebankWordDetokenizer().detokenize(tokens)
        return text
    numberTokens=len(tokens)
    try:
        if (numberTokens==0):
    #         start of sentence; use <s> & predict next one
            candidates=contextualDictionaryBooks[("<s>")]
        elif (numberTokens==1):
    #         add <s> at start of tokens
            candidates=contextualDictionaryBooks[("<s>",tokens[0])]
        elif ((tokens[numberTokens-2],tokens[numberTokens-1]) in contextualDictionaryBooks.keys()):
            candidates=contextualDictionaryBooks[(tokens[numberTokens-2],tokens[numberTokens-1])]            
        else:
            candidates=contextualDictionaryBooks[tokens[numberTokens-1]]
            
    except KeyError:
#         chosing random unigram we don't have much info related to context
        candidate="<s>"
        while (candidate in ["<s>","</s>"]):
            index=randrange(len(books_unigrams.keys()))
            candidate=list(books_unigrams.keys())[index]
        text+=" "+candidate
        return generateRandomSentenceBooks(text)
    #     now we need to chose from candidates
    try:
        sampleSize=10
        indicesToSample=random.sample(range(len(candidates)), sampleSize)
    except ValueError:
        sampleSize=len(candidates)
        indicesToSample=random.sample(range(len(candidates)), sampleSize)    
    
    index=indicesToSample[0]
    maxProb=candidates[index][1]
    for i in indicesToSample:
        if (candidates[index][1]>maxProb):
            index=i
#     index=randrange(len(candidates))
    text+=" "+candidates[index][0]
    return generateRandomSentenceBooks(text)
```

```python
def generateRandomSentenceMovies(text):
    maxTokens=100
    tokens=applySentenceTokenization(text)
 
    if (len(tokens)!=0 and (tokens[len(tokens)-1]=='.' or len(tokens)>maxTokens)):
        text=TreebankWordDetokenizer().detokenize(tokens)
        return text

    numberTokens=len(tokens)
    try:
        if (numberTokens==0):
            candidates=contextualDictionaryMovies[("<s>")]
        elif (numberTokens==1):
            candidates=contextualDictionaryMovies[("<s>",tokens[0])]

        elif ((tokens[numberTokens-2],tokens[numberTokens-1]) in contextualDictionaryMovies.keys()):
            candidates=contextualDictionaryMovies[(tokens[numberTokens-2],tokens[numberTokens-1])]            
        else:
#             print("in unigram context")
            candidates=contextualDictionaryMovies[tokens[numberTokens-1]]
    except KeyError:
        candidate="<s>"
        while (candidate in ["<s>","</s>"]):
            index=randrange(len(movies_unigrams.keys()))
            candidate=list(movies_unigrams.keys())[index]
        text+=" "+candidate
        return generateRandomSentenceMovies(text)

    try:
        sampleSize=10
        indicesToSample=random.sample(range(len(candidates)), sampleSize)
    except ValueError:
        sampleSize=len(candidates)
        indicesToSample=random.sample(range(len(candidates)), sampleSize)    
    
    index=indicesToSample[0]
    maxProb=candidates[index][1]
    for i in indicesToSample:
        if (candidates[index][1]>maxProb):
            index=i
#     index=randrange(len(candidates))
    text+=" "+candidates[index][0]
    return generateRandomSentenceMovies(text)
```
